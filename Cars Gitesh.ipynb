```python
import pandas as pd

import re

def get_year(text):
  pattern = r"\d{4}(?:[-\s]*[A-Za-z]+(?:\s*\d{2})?)?"
  match = re.search(pattern, text)
  return match.group(0) if match else ""  # Return the entire match (year)



df['Year_op'] = df['Year_of_Purchase'].apply(get_year)
print(df)
```

```python
import re

def extract_model_year(text):

  pattern = r"\d{4}"
  match = re.search(pattern, text)
  if match:
    return match.group(0)  # Return the captured year
  else:
    return None  # Return None if no year found

# Example usage
df['model_year'] = df['Name'].apply(extract_model_year)
print(df)
```

```python
import re

def extract_car_name(text):
  """Extracts the car name from a car description string.

  Args:
      text (str): The car description string (e.g., "2010 Maruti Alto LXI").

  Returns:
      str: The extracted car name (everything except the year), or None otherwise.
  """

  # This pattern assumes the year is at the beginning followed by a space
  pattern = r"\d+\s+(.*)"  
  match = re.search(pattern, text)
  if match:
    return match.group(1)  # Return everything captured after the space
  else:
    return None  # Return None if no match found

# Example usage
df['Car_name'] = df['Name'].apply(extract_car_name)
print(df)
```

```python
import pandas as pd
import re


def extract_km(text):

  pattern = r"\d+(,\d{3})*"  


  match = re.search(pattern, text)
  if match:
    kilometers = match.group(0).replace(",", "")  # Extract and remove commas
    return int(kilometers)  # Convert to integer
  else:

    if text.lower() == "kilometers km":
      return None  # Indicate invalid format
    else:
      return 0  

df['Kilometers'] = df['Kilometers'].apply(extract_km)

df['Kilometers'] = pd.to_numeric(df['Kilometers'], errors='coerce')  # Convert to numeric, replace errors with NaN

print(df)
```
